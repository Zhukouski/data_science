```python
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Предварительная обработка данных 
```

```python
df = df.drop(['ID'], axis=1)
```

```python
# Создаем экземпляр encoder'a
label_encoder = LabelEncoder()

# Тип дохода
df['Income_type'] = label_encoder.fit_transform(df['Income_type'])

# Уровень образования
df['Education_type'] = label_encoder.fit_transform(df['Education_type'])

# Семейное положение
df['Family_status'] = label_encoder.fit_transform(df['Family_status'])

# Тип жилья
df['Housing_type'] = label_encoder.fit_transform(df['Housing_type'])

# Сфера деятельности
df['Occupation_type'] = label_encoder.fit_transform(df['Occupation_type'])

# Просмотр результата
print(df.head())
```

```python
# Задача классификации 
```

```python
y=df['Target']
```

```python
X=df.drop(columns=['Target'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
X_trainx, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
X.shape
```

```python
# Decision Tree
```

```python
import gc
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
gc.collect() 
```

```python
model.fit(X_trainx,  y_train)
```

```python
pridictions = model.predict(X_test)
pridictions
```

```python
accuracy = accuracy_score(y_test, pridictions)
precision = precision_score(y_test, pridictions)
recall = recall_score(y_test, pridictions)
f1 = f1_score(y_test, pridictions)
    
print(f"Model: 'DecisionTreeClassifier'")
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-score: {f1:.2f}")
print()
```

```python
# Random Forest
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
model.fit(X_trainx,  y_train)
pridictions = model.predict(X_test)
accuracy = accuracy_score(y_test, pridictions)
precision = precision_score(y_test, pridictions)
recall = recall_score(y_test, pridictions)
f1 = f1_score(y_test, pridictions)

print(f"Model: 'RandomForestClassifier'")
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-score: {f1:.2f}")
print()
```

```python
# KNeighborsClassifier
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model =  KNeighborsClassifier(n_neighbors=6)
model.fit(X_trainx,  y_train)
pridictions = model.predict(X_test)
accuracy = accuracy_score(y_test, pridictions)
precision = precision_score(y_test, pridictions)
recall = recall_score(y_test, pridictions)
f1 = f1_score(y_test, pridictions)

print(f"Model: 'KNeighborsClassifier'")
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-score: {f1:.2f}")
print()
```
