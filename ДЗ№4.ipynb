```python
# Создаем экземпляр encoder'a
label_encoder = LabelEncoder()

# Месяц
df['Month'] = label_encoder.fit_transform(df['Month'])

# Тип посетителя
df['VisitorType'] = label_encoder.fit_transform(df['VisitorType'])

# Выходные дни
df['Weekend'] = label_encoder.fit_transform(df['Weekend'])

# Просмотр результата
print(df.head())
```

```python
# Задача классификации 
```

```python
y=df['Revenue']
```

```python
X=df.drop(columns=['Revenue'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
X_trainx, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
X.shape
```

```python
# Random Forest
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
model.fit(X_trainx,  y_train)
pridictions = model.predict(X_test)
accuracy = accuracy_score(y_test, pridictions)
precision = precision_score(y_test, pridictions)
recall = recall_score(y_test, pridictions)
f1 = f1_score(y_test, pridictions)

print(f"Model: 'RandomForestClassifier'")
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-score: {f1:.2f}")
print()
```
